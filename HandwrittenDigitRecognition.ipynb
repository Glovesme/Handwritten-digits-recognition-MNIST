### 1. KNN 

```python
import struct
import numpy as np
import operator
```

```python
def read_idx(filename):
    with open(filename, 'rb') as f:
        zero, data_type, dims = struct.unpack('>HBB', f.read(4))
        shape = tuple(struct.unpack('>I', f.read(4))[0] for d in range(dims))
        return np.frombuffer(f.read(), dtype=np.uint8).reshape(shape)
```

```python
'''KNN algorithm'''
def KNN(test_data, train_data, labels, k):
    train_data_size = train_data.shape[0]
    diffMatrix = np.tile(test_data, (train_data_size,1)) - train_data
    squareDiffMatrix = diffMatrix**2
    squareDistances = squareDiffMatrix.sum(axis = 1)
    distances = squareDistances**0.5
    sortedDistIndicies = distances.argsort()
    classCount={}
    for i in range(k):
        voteIlabel = labels[sortedDistIndicies[i]]
        classCount[voteIlabel] = classCount.get(voteIlabel, 0) + 1
    sortedClassCount = sorted(classCount.items(), key=operator.itemgetter(1), reverse=True)
    return sortedClassCount[0][0]
```

```python
def normalizeData(row, col, data):
    data.flags.writeable = True
    for m in range(row):
        for n in range(col):
            if data[m][n] > 0:
                data[m][n] = 1
    return data
```

```python
raw_train = read_idx('train-images.idx3-ubyte')
train_data = np.reshape(raw_train, (60000, 28*28))
train_lable = read_idx('train-labels.idx1-ubyte')
raw_test = read_idx('t10k-images.idx3-ubyte')
test_data = np.reshape(raw_test, (10000, 28*28))
test_label = read_idx('t10k-labels.idx1-ubyte')
```

```python
train_data = normalizeData(60000, 28*28, train_data)
test_data = normalizeData(10000, 28*28, test_data)
```

```python
test_data_size = test_data.shape[0]
```

```python
a = KNN(test_data[0], train_data, train_lable, 4)
a
```

```python
test_data_size = test_data.shape[0]
error_count = 0.0
for i in range(test_data_size):
    predictDigit = KNN(test_data[i], train_data, train_lable, 4)
    print('The actual digit is '+ str(test_label[i])+ 'what we predict is '+str(predictDigit))
    if predictDigit != test_label[i]:
        error_count += 1.0
print("\nThe total number of errors is: %d" % error_count)
print("\nThe total error rate is: %f" % (error_count/test_data_size))
    
```
